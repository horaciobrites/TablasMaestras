## Nuevos Registros

```python
import pandas as pd
import numpy as np
# ================== Funciones ================================
def parse_numero(serie):
    return (
        serie
        .astype(str)
        .str.replace('.', '', regex=False)   # elimina separador de miles
        .str.replace(',', '.', regex=False)   # coma → punto decimal
        .pipe(pd.to_numeric, errors='coerce')
    )
# ============================ PM_Aserradero2 =============================
df1 = pd.read_csv(r"C:\Users\horacio.brites\OneDrive - ARAUCO\DW Qlik\DataExcels\Maestro de Maeriales.txt",sep="\t",encoding="latin-1",
    skiprows=3,          # saltea título + línea en blanco + encabezado de columnas vacío
    header=0,            # la fila 4 (índice 3) es el header real
    skipinitialspace=True,
    dtype=str            # todo como string primero, para no perder nada
)

df1.columns = df1.columns.str.strip()
df1 = df1.dropna(subset=["Objeto"])
df1["Objeto"] = df1["Objeto"].str.strip()
df1 = df1[df1["Objeto"] != ""]
df1=df1[['Objeto', 'Denomin.', 'Tipo de madera', 'Familia','Calidad', 'Terminación','Grado', 'Espesor', 'Ancho', 'Largo']]
print(df1[df1['Objeto']=="1524107"])

#=========================== PM_Aserradero ===============================================
df2 = pd.read_csv(r"C:\Users\horacio.brites\OneDrive - ARAUCO\DW Qlik\DataExcels\Maestro de Maeriales2.txt",sep="\t",encoding="latin-1",
    skiprows=3,          # saltea título + línea en blanco + encabezado de columnas vacío
    header=0,            # la fila 4 (índice 3) es el header real
    skipinitialspace=True,
    dtype=str            # todo como string primero, para no perder nada
)
df2.columns = df2.columns.str.strip()
df2 = df2.dropna(subset=["Objeto"])
df2["Objeto"] = df2["Objeto"].str.strip()
df2 = df2[df2["Objeto"] != ""]
df2=df2[['Objeto', 'Denomin.', 'Tipo de madera', 'Familia','Calidad', 'Terminación','Grado', 'Espesor', 'Ancho', 'Largo']]
#=================================== Blanks =========================================
df3 = pd.read_csv(
    r"C:\Users\horacio.brites\OneDrive - ARAUCO\DW Qlik\DataExcels\Maestro de Materiales_blanks.txt",
    sep="\t",
    encoding="latin-1",
    skiprows=3,
    dtype=str
)
df3.columns = df3.columns.str.strip()
df3 = df3.dropna(subset=["Objeto"])
df3["Objeto"] = df3["Objeto"].str.strip()
df3 = df3[df3["Objeto"] != ""]
df3=df3[['Objeto', 'Denomin.', 'Tipo de madera', 'Familia','Calidad', 'Terminación','Grado', 'Espesor', 'Ancho', 'Largo']]
#======================================= Blocks ==========================================
df4 = pd.read_csv(
    r"C:\Users\horacio.brites\OneDrive - ARAUCO\DW Qlik\DataExcels\Maestro de Materiales_block.txt",
    sep="\t",
    encoding="latin-1",
    skiprows=3,
    dtype=str
)
df4.columns = df4.columns.str.strip()
df4 = df4.dropna(subset=["Objeto"])
df4["Objeto"] = df4["Objeto"].str.strip()
df4 = df4[df4["Objeto"] != ""]
df4=df4[['Objeto', 'Denomin.', 'Tipo de madera', 'Familia','Calidad', 'Terminación','Grado', 'Espesor', 'Ancho', 'Largo']]

#================================== df_base ===============================
df_con = pd.concat([df1, df2, df3, df4], ignore_index=True)
df_con["Terminación"] = df_con["Terminación"].replace({"RG": "RGH"})
df_con['Ancho'] = df_con['Ancho'].fillna(0)
df_base = pd.read_csv(r"C:\Users\horacio.brites\OneDrive - ARAUCO\DW Qlik\DataExcels\BDBase.csv",encoding="utf-8",dtype=str)
df_nuevos = df_con[~df_con["Objeto"].isin(df_base["Objeto"])].copy()
df_nuevos[['Espesor', 'Ancho', 'Largo']] = df_nuevos[['Espesor', 'Ancho', 'Largo']].apply(parse_numero)
#df_nuevos[['Espesor', 'Ancho', 'Largo']] = (df_nuevos[['Espesor', 'Ancho', 'Largo']].replace(',', '.', regex=True).apply(pd.to_numeric, errors='coerce'))
df_nuevos

```

## Divide los materiales en PP y PT

```python
df_nuevos_PT=df_nuevos[df_nuevos['Denomin.'].str.startswith("PT")].copy()
df_nuevos_PP=df_nuevos[df_nuevos['Denomin.'].str.startswith("PP")].copy()
df_nuevos_PT
```

## Codificiacion de productos terminados

```python
#Materiales que comienzan con PT
#1- EspesorPPCalc
df_nuevos_PT.loc[:, "EspesorPPCalc"] = np.where(
    df_nuevos_PT["Terminación"] == "RGH",
    round(df_nuevos_PT["Espesor"] * (1 + 0.035), 0),

    np.where(
        df_nuevos_PT["Terminación"].isin(["S2", "S4", "R4"]),
        round((df_nuevos_PT["Espesor"] + 2*1.5 + 0.2*2) * (1 + 0.035), 0),

        np.where(
            df_nuevos_PT["Terminación"].isin(["S1", "S3"]),
            round((df_nuevos_PT["Espesor"] + 1.5 + 0.2*1) * (1 + 0.035), 0),
            np.nan
        )
    )
)


#2- AnchoPPCalc
df_nuevos_PT.loc[:, "AnchoPPCalc"] = np.where(
    (df_nuevos_PT["Ancho"] == 0) | (df_nuevos_PT["Ancho"].isna()),
    0,

    np.where(
        (df_nuevos_PT["Terminación"] == "RGH") & (df_nuevos_PT["Ancho"] > 120),
        round(df_nuevos_PT["Ancho"] * (1 + 0.04), 0),

        np.where(
            df_nuevos_PT["Terminación"].isin(["RGH", "S1", "S2"]),
            round(df_nuevos_PT["Ancho"] * (1 + 0.035), 0),

            np.where(
                df_nuevos_PT["Terminación"].isin(["S3", "S4", "R4"]),
                round((df_nuevos_PT["Ancho"] + 2 * 1.5 + 0.2 * 2) * (1 + 0.035), 0),
                np.nan
            )
        )
    )
)

#3-LargoPPCalc


map_largo = {
    1014: 3050,
    936: 3050,
    1217: 3660,
    1000: 3050,
    1100: 3355,
    1200: 3660,
    1218: 3660,
    915: 3050,
    1220: 3660,
    1118: 4575
}

df_nuevos_PT["Largo"] = pd.to_numeric(df_nuevos_PT["Largo"], errors="coerce")

df_nuevos_PT["LargoPPCalc"] = np.where(
    df_nuevos_PT["Familia"].isin(["BK", "MU"]),
    0,

    np.where(
        df_nuevos_PT["Familia"] == "CP",
        df_nuevos_PT["Largo"].map(map_largo),

        np.where(
            (df_nuevos_PT["Tipo de madera"] == "C") & (df_nuevos_PT["Largo"] <= 2440),
            df_nuevos_PT["Largo"] * 2,
            df_nuevos_PT["Largo"]
        )
    )
)

#4-Kerf
df_nuevos_PT.loc[:, "Kerf"] = np.where(
    (df_nuevos_PT["Tipo de madera"] == "C") & (df_nuevos_PT["Largo"] == 1830),
    "6+6",
    np.where(
        (df_nuevos_PT["Tipo de madera"] == "C") & (df_nuevos_PT["Largo"] == 2440),
        "8+8",
        np.where(
            df_nuevos_PT["Familia"] == "CP",
            np.where(
                df_nuevos_PT["Largo"] >= 1200,
                "4+4+4",
                np.where(
                    df_nuevos_PT["Largo"] == 1100,
                    "",
                    np.where(
                        df_nuevos_PT["Largo"] == 1118,
                        "",
                        "3+3+3"
                    )
                )
            ),
            ""
        )
    )
)


#RendVolCalc
df_nuevos_PT["RendVolCalc"] = np.where(
    df_nuevos_PT["Familia"] == "CP",
    (
        df_nuevos_PT["Espesor"] * df_nuevos_PT["Ancho"] * df_nuevos_PT["Largo"] *
        (df_nuevos_PT["LargoPPCalc"] / df_nuevos_PT["Largo"]).round(0)
    ) / (
        df_nuevos_PT["EspesorPPCalc"] * df_nuevos_PT["AnchoPPCalc"] * df_nuevos_PT["LargoPPCalc"]
    ),
    (
        df_nuevos_PT["Espesor"] * df_nuevos_PT["Ancho"]
    ) / (
        df_nuevos_PT["EspesorPPCalc"] * df_nuevos_PT["AnchoPPCalc"]
    )
)


# 7- CodigoPP

esp = df_nuevos_PT["EspesorPPCalc"].astype(int).astype(str)
anc = df_nuevos_PT["AnchoPPCalc"].astype(int).astype(str)
lar = df_nuevos_PT["LargoPPCalc"].astype(int).astype(str)
kerf = df_nuevos_PT["Kerf"].astype(str)

base = esp + anc + lar + kerf

df_nuevos_PT["CodigoPP"] = np.select(
    [
        df_nuevos_PT["Familia"] == "BK",
        df_nuevos_PT["Familia"] == "PF"
    ],
    [
        "RP" + base,
        "MA" + base
    ],
    default=df_nuevos_PT["Familia"].astype(str) + base
)

# 8- PrecodPP
base2 = esp + anc
df_nuevos_PT["PrecodPP"] = np.select(
    [
        df_nuevos_PT["Familia"] == "BK",
        df_nuevos_PT["Familia"] == "PF"
    ],
    [
        "RP" + base2,
        "MA" + base2
    ],
    default=df_nuevos_PT["Familia"].astype(str) + base2
)

# 9 - Escuadria
df_nuevos_PT.loc[:, "Escuadria"] = (
    df_nuevos_PT["Familia"].astype(str)
    + " " + df_nuevos_PT["EspesorPPCalc"].astype(int).astype(str)+"x"
    + df_nuevos_PT["AnchoPPCalc"].astype(int).astype(str)
)


# 10 - Calidad
map_calidad = {
    "PA": "Pallet",
    "PR": "Premium",
    "ST": "Standard",
    "SP": "Standard Plus",
    "PM": "Pallet",
    "MA": "Machimbre",
    "EN": "G2 Standard Plus",
    "AE": "Excedente",
    "EM": "Excedente",
    "CP": "Comp. Pallet",
    "CH": "Comp. Pallet",
    "CO":"COL"
}

df_nuevos_PT["Calidad2"] = np.where(
    df_nuevos_PT["Familia"] == "BK",
    "Blank",

    np.where(
        df_nuevos_PT["Familia"] == "MU",
        "COL",

        np.where(
            (df_nuevos_PT["Familia"] == "CP") | (df_nuevos_PT["Objeto"] == "1524863"),
            "Comp. Pallet",
            df_nuevos_PT["Calidad"].map(map_calidad).fillna("ver")
        )
    )
)


#11- CodAgrup
df_nuevos_PT["CodAgrup"] = np.where(
    (df_nuevos_PT["Familia"] == "MU") & (df_nuevos_PT["Terminación"] == "RGH"),
    df_nuevos_PT["Espesor"].astype(int).astype(str),
    df_nuevos_PT["Objeto"]
)

df_nuevos_PT = df_nuevos_PT.drop(columns=["Calidad"])
df_nuevos_PT = df_nuevos_PT.rename(columns={"Calidad2": "Calidad"})
#12 Familia Postiza
Map = {
    'PF': 'MA',
    'BK': 'RP',
    'BL': 'RP',
    'MA': 'MA',
    'CP': 'CP'
}

df_nuevos_PT.loc[:, "FamiliaPostiza"] = np.where(
    df_nuevos_PT["Familia"] == "MU",
    np.where(
        df_nuevos_PT["Terminación"] == "RGH",
        "MU",
        "MUX"
    ),
    df_nuevos_PT["Familia"].map(Map)
)
df_nuevos_PT=df_nuevos_PT[['Objeto', 'Denomin.', 'Tipo de madera', 'Familia', 'Calidad',
       'Terminación', 'Grado', 'Espesor', 'Ancho', 'Largo', 'EspesorPPCalc',
       'AnchoPPCalc', 'LargoPPCalc', 'Kerf', 'RendVolCalc',
       'CodigoPP', 'PrecodPP', 'Escuadria','CodAgrup',"FamiliaPostiza"]]
#12- Status
df_nuevos_PT['Status']=True

df_nuevos_PT
```
